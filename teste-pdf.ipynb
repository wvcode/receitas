```python
import pdfplumber
import pandas as pd
import re
import os
import pprint
```

```python
pdf = pdfplumber.open(os.path.join("dados", "content.pdf"))

data = []
start_parse = False

for num, page in enumerate(pdf.pages):
    rows = page.extract_text_lines()
    for item in rows:
        data.append(item["text"])
pdf.close()


titulos = []
for idx, item in enumerate(data):
    if item in ["Ingredientes", "Ingredientes:", "Massa:"]:
        titulos.append(data[idx - 1])


# pprint.pprint(data, indent=4)
pprint.pprint(titulos, indent=4)
```
